```python

from pythonosc import udp_client
import random
import time
```

```python


# Create OSC client
client = udp_client.SimpleUDPClient("127.0.0.1", 5005)  # Change IP/port as needed
count=0
# Main loop
while True:
    # Generate random number between 0 and 100
    random_num = random.randint(0, 100)
    
    # Send the random number via OSC
    client.send_message("/random", count)
    print(f"Sent: {count}")
    count+=1
    # Wait 3 seconds
    time.sleep(1)
```
